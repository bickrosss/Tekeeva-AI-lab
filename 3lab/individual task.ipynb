```python
import matplotlib.pyplot as plt
import numpy as np
months = ['', '', '', '', '', '', 
          '', '', '', '', '', '']
precipitation = [45, 38, 50, 60, 70, 90, 110, 95, 80, 65, 55, 48]
average = np.mean(precipitation)
plt.figure(figsize=(10, 5))
plt.title("   ")
plt.xlabel("")
plt.ylabel(" ()")
plt.grid()
for i in range(len(months)):
    if precipitation[i] >= average:
        plt.plot(months[i], precipitation[i], 'bo')
    else:
        plt.plot(months[i], precipitation[i], 'ro')
plt.plot(months, precipitation, 'k-', alpha=0.3)

plt.axhline(y=average, color='g', linestyle='--', 
            label=f': {average:.1f} ')
plt.legend()

plt.show()
```

```python
countries = ['', '', '', '', '']
ages = [84, 83, 82, 81, 78]
colors = ['green' if x >= 80 else 'orange' for x in ages]

plt.barh(countries, ages, color=colors)
plt.axvline(x=80, color='red', linestyle='--')
plt.xlabel(' ()')
plt.title('  ')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 200)
y = 2*np.log(x) + x

print(":", np.trapz(y, x))

plt.plot(x, y, 'g')
plt.fill_between(x, y, color='lightgreen')
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
t = np.linspace(0, 4*np.pi, 500)
x = t * np.cos(t)
y = t * np.sin(t)
z = t
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, linewidth=2, color='green')
ax.set_title('3D  ', fontsize=14)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.tight_layout()
plt.show()
```
