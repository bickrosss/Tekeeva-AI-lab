```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 200)
plt.plot(x, x**2)
plt.title('График y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
x = np.linspace(-15, 15, 200)
plt.figure(figsize=(20, 10))

plt.axis([-7, 7, -7, 7])

plt.plot(x, x, "--b", linewidth=2, label=r"$y = x$")
plt.plot(x, x**2, "-.g", linewidth=2, label=r"$y = x^2$")
plt.plot(x, x**3, "-r", linewidth=2, label=r"$y = x^3$")

plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.random.randint(-10, 10, 50)  
y = np.random.randint(-10, 10, 50)  
plt.scatter(x, y, c=x, s=abs(y)*10, cmap='summer')
plt.colorbar(label='Значение x')
plt.title('Диаграмма рассеяния')
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, color='tan')
plt.axvline(np.mean(data), color='g', linestyle='--')
plt.title('Гистограмма распределения')
plt.xlabel("Значение")
plt.ylabel("Частота")
```

```python
marks = ["Отлично", "Хорошо", "Удовлетворительно", "Неудовлетворительно"]
amount = [15, 30, 25, 20]
plt.bar(marks, amount, color=["limegreen", "forestgreen", "darkgreen", "seagreen"])
plt.xlabel("Оценки")
plt.ylabel("Количество студентов")
plt.title("Оценки студентов")
```

```python
marks = ["Отлично", "Хорошо", "Удовлетворительно", "Неудовлетворительно"]
amount = [20, 35, 30, 15]
plt.pie(amount, labels=marks, autopct="%1.1f%%", colors=["limegreen", "forestgreen", "darkgreen", "seagreen"])
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='summer')
plt.title('3D поверхность')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)
z = np.linspace(0, 2*np.pi, 100)

axs[0,0].plot(x, x)
axs[0,0].set_title('y = x')

axs[0,1].plot(x, x**2)
axs[0,1].set_title('y = x²')

axs[1,0].plot(z, np.sin(z))
axs[1,0].set_title('y = sin(z)')

axs[1,1].plot(z, np.cos(z))
axs[1,1].set_title('y = cos(z)')

plt.tight_layout()
```
