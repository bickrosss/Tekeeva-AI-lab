```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range('2024-11-15', periods=24, freq='h')
defects = np.random.randint(0, 11, size=len(dates))
production = pd.Series(defects, index=dates)

outliers = production[production > 8]

plt.figure(figsize=(10, 5))
production.plot(color='green', marker='o', label='Дефекты')
plt.scatter(outliers.index, outliers, 
            color='orange', s=100, label='Выбросы (>8)')
plt.title('Дефектные изделия 15.11.2024')
plt.ylabel('Количество')
plt.legend()
plt.grid()
plt.show()
```
